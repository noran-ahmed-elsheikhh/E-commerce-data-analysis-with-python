```python
# Filling NaN for numerics with mean , categorical with mode
Numeric_cols = DF.select_dtypes(include=['int64','float64']).columns
Categorical_cols = DF.select_dtypes(include=['object']).columns

DF[Numeric_cols] = DF[Numeric_cols].fillna(DF[Numeric_cols].mean())
DF[Categorical_cols] = DF[Categorical_cols].fillna(DF[Categorical_cols].mode())
```

```python
print(DF[Numeric_cols])
print(DF[Categorical_cols])
```

```python
# making sure numeric cols are int or float
DF["quantity"] = pd.to_numeric(DF["quantity"], errors="coerce")
DF["unit_price"] = pd.to_numeric(DF["unit_price"], errors="coerce")
DF["discount"] = pd.to_numeric(DF["discount"], errors="coerce")
DF["shipping_cost"] = pd.to_numeric(DF["shipping_cost"], errors="coerce")
DF["total_sales"] = pd.to_numeric(DF["total_sales"], errors="coerce")
```

```python
# removing negative numbers from numeric cols
DF["quantity"] = DF["quantity"].abs()
DF["unit_price"] = DF["unit_price"].abs()
DF["discount"] = DF["discount"].abs()
DF["shipping_cost"] = DF["shipping_cost"].abs()
DF["total_sales"] = DF["total_sales"].abs()
```

```python
DF = DF.drop_duplicates()
print(DF.head())
```

```python
# converting Dates cols into datetime
DF['delivery_date'] = pd.to_datetime(DF['delivery_date'],errors='coerce')
DF['order_date'] = pd.to_datetime(DF['order_date'],errors='coerce')
DF['ship_date'] = pd.to_datetime(DF['ship_date'],errors='coerce')

DF['delivery_date']= DF['delivery_date'].dt.date
DF['order_date']= DF['order_date'].dt.date
DF['ship_date']= DF['ship_date'].dt.date

DF['Total Revenue'] = DF['quantity'] * DF['unit_price']
print(DF['Total Revenue'])

DF['Total Cost'] = DF['shipping_cost'] + DF['discount']
print(DF['Total Cost'])

DF['Profit'] = DF['Total Revenue'] + DF['Total Cost']
print(DF['Profit'])

def Profit_Classification(p):
  if p > 0 :
    return "Positive +"
  elif p < 0:
    return "Negative -"
  else:
    return "Zero"

DF['Profit_Class'] = DF['Profit'].apply(Profit_Classification)
print(DF['Profit_Class'])
```
